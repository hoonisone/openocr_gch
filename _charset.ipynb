```python
from typing import List, Set
from pathlib import Path


def load_charset(charset_file_path:str|Path)->Set:
    with open(charset_file_path) as f:
        charset = [line.strip() for line in f]

    return set(charset)

def load_charset_from_label_file(label_file_path:str|Path)->Set:
    charset = set()
    with open(label_file_path) as f:
        for line in f:
            path, text = line.strip().split('\t')
            charset.update(set(list(text)))
    return charset
```

```python
charset_path = "/home/resources/gch/2_charsets/1_single/4_korean/1_character/ppocr_core___id_6/charset.txt"
aihub_path = "/home/resources/shared_datasets/STR/aihub_rec/clean_label.txt"

charset = load_charset(charset_path)
aihub_charset = load_charset_from_label_file(aihub_path)

print(len(charset))
print(len(aihub_charset))

print(len(charset-aihub_charset))
print(len(aihub_charset-charset))
```
