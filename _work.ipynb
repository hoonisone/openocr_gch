```python
work_ids = [52, 53, 54, 55]
work_ids = [56, 57, 58, 59]

work_ids = [56, 57, 58, 59, 60]
work_ids = [62]

gpu_ids = [0 , 1, 2, 3]
accumalate = False

commands = []

for work_id in work_ids:
    work_context = factory.get_work_context(work_id)
    train_task = work_context.get_train_task_record()
    train_context = work_context.get_train_task_context(train_task)
    train_command = train_context.make_train_command()
    commands.append(train_command)

factory.get_command_manager().write_commands("train", gpu_ids, commands, accumalate=accumalate)
```
